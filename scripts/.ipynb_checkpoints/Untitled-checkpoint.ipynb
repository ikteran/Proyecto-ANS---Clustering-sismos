```python
# df_velo= pd.read_excel(data_velo, header=0)
data_stations = pd.read_excel('TimeStation.xlsx', header=0)
x = data_stations.long
y = data_stations.lat
print(data_stations)
```

```python
r = [-81, -66, -5, 15]
p = 'M0/12c'
f =["SenW","a5f"]

topo_data = pygmt.datasets.load_earth_relief(resolution="30s", region=r)

pygmt.makecpt(
    cmap="gray91,dimgray", 
    # Elegir los colores de la paleta. Pueden ser los que se encuentran  
    # en la tabla de colores de GMT o se puede ingresar el serial RGB del color.
    continuous=False,
    background=True,
    output="dodger_gray.cpt"
    # Nombrar como se va a denominar la paleta. Tiene que llevar 
    # la extensión .cpt
)

fig = pygmt.Figure()

# CONFIGURATION MAP
pygmt.config(FONT_ANNOT_PRIMARY= "8p,Helvetica,black")
pygmt.config(MAP_FRAME_TYPE= "inside") #lines

pygmt.makecpt(
cmap='dodger_gray.cpt',
reverse=True,
# series='0/1/0.001',
continuous=True
)


fig.grdimage(
grid=topo_data,
region=r,
projection=p,
shading=True,
frame=f
)


fig.basemap(region=r, 
    projection=p, 
    frame=f)

fig.coast(
    region=r, 
    projection=p,
    frame=f,
    water="gray77",
    borders = ["1/0.3p,gray32"], #Countries
    shorelines= ["1/0.7p,gray44"],
    resolution="fa"
)

# ADD FAULTS
# Fallas principales
fig.plot(
    data=fallas_pples,
    pen="0.6p,black", 
    label='"Falla"'
)

# Fallas Secundarias
fig.plot(
    data=fallas_sec,
    pen="0.5p,black",
    transparency=20, 
)

# Zonas de subducción1
fig.plot(
    data=zon_sub1,
    pen="0.6p,black",
    style="f0.7c/0.13c+r+t",
    color="black"
)

# Zonas de subducción2
fig.plot(
    data=zon_sub2,
    pen="0.6p,black",
    style="f0.7c/0.13c+r+t",
    # color="white"
)

# ADD PLATES LABELS
fig.text(
    textfiles="plates.txt", 
    angle=True, 
    font=True, 
    justify=True,
    fill="white",
    transparency=20
)

# ADD COUNTRIES LABELS
# fig.text(
#     textfiles="countries.txt", 
#     angle=True, 
#     font=True,
#     fill="white",
#     justify=True,
#     transparency=30
# )

# #ADD RECTANGLE TO VELOCITY LEGEND SAME SIZE TO COLORBAR
# fig.plot(x=-70.4, y=-0.1, 
#          style="r3.8/0.8", 
#          color="white", 
#          pen="0.5p,black")

# #ADD RECTANGLE TO VELOCITY LEGEND 
# fig.plot(x=-68.34, y=-0.1, 
# fig.plot(x=-68.34, y=0.15, 
#           style="r2.1/1.0", 
#           color="white", 
#           pen="0.5p,black")

# ADD VECTOR LEGEND
# fig.text(
#     textfiles="vector.txt", 
#     angle=True, 
#     font=True, 
#     justify=True,
#     )

# ADD ID STATIONS LEGEND
# fig.text(
#     textfiles="ID_P6.txt", 
#     angle=True, 
#     font=True, 
#     justify=True,
#     )

# ADD FAULTS LEGEND
pygmt.config(FONT_ANNOT_PRIMARY= "7p,Helvetica,black")
fig.legend(
    position="g-68.75/-1.8",
    box="+gwhite+p0.5p",
)


# fig.coast(land="black", water="skyblue")
# pygmt.makecpt(cmap="lajolla", series=[data_stations.tim.min(), data_stations.tim.max()])
pygmt.makecpt(cmap="lajolla", series=[2.5, data_stations.tim.max()])


# fig.velo(
#     data=df_velo,
#     region=r,
#     pen="0.9p,brown",
#     line=True,
#     spec="e0.03/0/5", #sizevectors/confidece/sizelabels
#     # frame=["WSne", "2g2f"], #title
#     projection=p,
#     vector="0.15c+p0.5p+e+gbrown+h0.5",
#     resolution="h",
#     label="Vector"
#     )

# #ID
# fig.velo(
#     data=df_velo,
#     region=r,
#     pen="0.9p,brown",
#     line=True,
#     spec="e0.28/0/5", #sizevectors/confidece/sizelabels
#     # frame=["WSne", "2g2f"], #title
#     projection=p,
#     vector="0.2c+p0.5p+e+gbrown+h0.5",
#     resolution="h",
#     label="Vector"
#     )


fig.plot(x=x, y=y,
    # size="c0.02",S
    color=data_stations.tim / data_stations.tim.max(),
    cmap="lajolla",
    style="c0.18c",
    pen="black",
)

# ADD COLORBAR
pygmt.config(FONT_LABEL= "8p,Helvetica,black")
pygmt.config(FONT_ANNOT_PRIMARY= "7p,Helvetica,black")
pygmt.config(MAP_FRAME_TYPE= "fancy")
fig.colorbar(
    position="g-73.4/-2.6+w5c/0.20c+h",
    box="+gwhite+p0.5p",
    frame=['"+LTiempo de datos(años)"', "a2"]
    )

fig.savefig("time_stations.png",
            dpi=1000,
            )
fig.show()
# Añadir una paleta de colores al mapa
```
